```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
from d_d_n_he3 import *  
```

```python
ax = plt.axes(projection="3d")
energies = np.logspace(4.1, 8, num=100)
angles = np.linspace(0, np.pi, 100)
xv, yv = np.meshgrid(np.log(energies), angles)

z2 = np.zeros((len(angles), len(energies)))
for i in range(len(energies)):
    e = energies[i]
    for j in range(len(angles)):
        angle = angles[j]
        z2[j, i] = max(diff_cross_2(e, angle), 1)
        

ax.set_title("Differential cross section $D(D,n)^3 He$")
ax.set_xlabel("$ \mathcal{E}$ (eV/u)")
ax.set_ylabel("$ \Theta $")    
ax.set_zlabel("$ d\sigma/d\Omega$ (mb/sr)") 
ax.plot_surface(xv, yv, np.log(z2), color=(0,0,1,0.8))

plt.show()
```
